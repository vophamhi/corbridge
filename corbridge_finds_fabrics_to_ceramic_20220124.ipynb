```python
df_result = df
counter = 0
for ind in df_result.index:
    if isinstance(df_result['Fabric'][ind], str):
        if not isinstance(df_result['Material'][ind], str):
            df['Material'][ind] = 'ceramic'
            print(ind)
            print(df_result['Fabric'][ind])
            print(df_result['Material'][ind])
            print('----------')
            counter += 1
```
